```python
# Assume the dataset "India_budget_2021.csv" is already loaded in data
data = pd.read_csv("India_budget_2021.csv")

# Select specific rows
data = data.iloc[[0, 8, 11, 14, 18, 23, 41, 42, 43], :]

# Create a new row as a DataFrame
row = pd.DataFrame([{'Department /Ministry': 'OTHERS', 'Fund allotted(in ₹crores)': 592971.0800000001}])

# Concatenate the new row to the selected data
data = pd.concat([data, row], ignore_index=True)

# Print the resulting DataFrame
print(data)
```

```python
data.plot.bar(x='Department /Ministry', y='Fund allotted(in ₹crores)')
```

```python
df = data["Fund allotted(in ₹crores)"]
labels = data["Department /Ministry"]
plt.figure(figsize=(12,12))
plt.pie(df, labels=labels, autopct='%1.2f%%', startangle=90, pctdistance=0.85, shadow =True)
central_circle = plt.Circle((0, 0), 0.4, color='white')
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font', size=12)
plt.title("Distribution of The Budget", fontsize=20)
plt.show()
```
